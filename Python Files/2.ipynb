```python
import tensorflow as tf  # Importing TensorFlow library for deep learning operations
import numpy as np  # Importing NumPy library for numerical operations
from sklearn.metrics import classification_report,confusion_matrix  # Importing classification_report from scikit-learn library
model = tf.keras.models.load_model("model_1.keras")  # Loading the pre-trained model
```

```python
def load_data():
    x_train = np.load("x_train.npy")  # Loading training images
    y_train = np.load("y_train.npy")  # Loading training labels
    x_test = np.load("x_test.npy")  # Loading testing images
    y_test = np.load("y_test.npy")  # Loading testing labels
    
    return x_train, y_train, x_test, y_test
```

```python
x_train, y_train, x_test, y_test = load_data()  # Loading data
x_test = x_test.astype('float32') / 255.0  # Normalizing testing images
y_test.reshape(y_test.shape[0], 1)  # Reshaping testing labels
```

```python
# evaluate model
#print(f"Accuracy: {model.evaluate(x_test, y_test)[1] * 100}%")
#print(model.summary())
```

```python
predictions = model.predict(x_test)  # Making predictions on testing data
predictions[predictions >= 0.5] = 1  # Thresholding predictions
predictions[predictions < 0.5] = 0
```

```python
#print(predictions)  # Printing the predicted labels
print(classification_report(y_test, predictions))  # Printing classification report
#tf.keras.utils.plot_model(model, to_file="C:/Users/vedan/Desktop/NUV/Tyre_Classification/50_epochs_architecture.png", show_shapes=True)  # Plotting the model architecture and saving it as an image
```
